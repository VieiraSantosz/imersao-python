```python
# bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import mplfinance as mpf
import yfinance as yf 
```

```python
# pela a biblioteca da yfinance, pegar a bolsa de valores da Petrobras no ano de 2023
df_petrobas = yf.download('PETR4.SA', start='2023-01-01', end='2023-12-31')

display(df_petrobas)
```

```python
# renomear todas as colunas do DataFrame
df_petrobas.columns = ['Abertura', 'Maximo', 'Minimo', 'Fechamento', 'Fech_Ajust', 'Volume']

display(df_petrobas.columns)
```

```python
# renomear a parte do Date para Data
df_petrobas = df_petrobas.rename_axis('Data')

display(df_petrobas)
```

```python
# criaÃ§Ã£o de um grÃ¡fico com a variaÃ§Ã£o dos preÃ§os por data
df_petrobas['Fechamento'].plot(figsize=[12, 8])

plt.title('VariaÃ§Ã£o do PreÃ§o por Data', fontsize=16)

plt.legend(['Fechamento'])
```

```python
# criaÃ§Ã£o de um Candlestick das aÃ§Ãµes da Petrobras
dados = yf.download('PETR4.SA', start='2023-01-01', end='2023-12-31')

mpf.plot(dados.head(60), type='candle', figsize=(16, 8),
        volume=True, mav=(7, 14), style='yahoo')
```
